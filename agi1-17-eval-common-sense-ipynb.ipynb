```python
# AGI1_17_eval_common_sense.ipynb
# Evaluate AGI on common-sense Q&A

!pip install transformers --quiet
from transformers import pipeline

# Load simple LLM
common_sense_agent = pipeline("text-generation", model="distilgpt2")

# Sample common-sense questions
questions = [
    "You drop a glass on the floor. What happens?",
    "If you go outside during a rainstorm without an umbrella, what happens?",
    "Can a dog drive a car?",
    "Why do people wear jackets in winter?",
    "Is it okay to put a phone in the oven?",
]

# Run the test
for q in questions:
    print(f"\n🌍 Question: {q}")
    prompt = f"Q: {q}\nA:"
    reply = common_sense_agent(prompt, max_length=60, do_sample=False)[0]["generated_text"]
    print(f"✅ AGI Common Sense Answer:\n{reply.strip()}")
```
