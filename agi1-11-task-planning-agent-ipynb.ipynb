```python
# AGI1_11_task_planning_agent.ipynb
# Convert high-level instruction into task steps using GPT-2

!pip install transformers --quiet
from transformers import pipeline

# âœ… Load GPT-2 small model
planner = pipeline("text-generation", model="gpt2")

# âœ… Define user goal (can be voice or text input)
goal = "Analyze customer support calls and generate a weekly summary."

# âœ… Prompt for planning
prompt = f"""
User Goal: {goal}
Break it into step-by-step tasks:

1.
"""

# âœ… Generate plan
plan = planner(prompt, max_length=100, do_sample=False)[0]['generated_text']

print("ðŸ§  Task Planning Agent Output:\n")
print(plan)
```
