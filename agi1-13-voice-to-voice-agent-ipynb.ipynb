```python
# AGI1_13_voice_to_voice_agent.ipynb
# Simulate voice → text → LLM → text (TTS not supported on Kaggle)

!pip install transformers torchaudio --quiet

from transformers import WhisperProcessor, WhisperForConditionalGeneration, pipeline
import torchaudio

# ✅ Load Whisper model
whisper_model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-tiny.en")
whisper_processor = WhisperProcessor.from_pretrained("openai/whisper-tiny.en")

# ✅ Load LLM
llm = pipeline("text-generation", model="distilgpt2")

# ✅ Load voice input (you can replace this with your own file later)
sample_audio, sr = torchaudio.load("https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/sample1.flac")

# ✅ Step 1: Transcribe
input_features = whisper_processor(sample_audio[0], sampling_rate=sr, return_tensors="pt").input_features
generated_ids = whisper_model.generate(input_features)
transcribed_text = whisper_processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print("🎧 Transcribed voice input:\n", transcribed_text)

# ✅ Step 2: Generate response
prompt = f"User said: '{transcribed_text}'\nAGI reply:"
response = llm(prompt, max_length=50, do_sample=True)[0]['generated_text']

print("\n🗣️ AGI Voice Agent Reply:\n", response)
```

```python
# AGI1_13_voice_to_voice_agent.ipynb
!pip install transformers torchaudio --quiet

from transformers import WhisperProcessor, WhisperForConditionalGeneration, pipeline
import torchaudio
import requests

# ✅ Step 1: Download audio from Hugging Face to file
url = "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/sample1.flac"
r = requests.get(url)
with open("sample1.flac", "wb") as f:
    f.write(r.content)

# ✅ Step 2: Load it with torchaudio
sample_audio, sr = torchaudio.load("sample1.flac")

# ✅ Step 3: Load Whisper
whisper_model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-tiny.en")
whisper_processor = WhisperProcessor.from_pretrained("openai/whisper-tiny.en")

# ✅ Step 4: Transcribe voice
input_features = whisper_processor(sample_audio[0], sampling_rate=sr, return_tensors="pt").input_features
generated_ids = whisper_model.generate(input_features)
transcribed_text = whisper_processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
print("🎧 Transcribed Voice:\n", transcribed_text)

# ✅ Step 5: Load LLM and reply
llm = pipeline("text-generation", model="distilgpt2")
prompt = f"User said: '{transcribed_text}'\nAGI reply:"
response = llm(prompt, max_length=50, do_sample=True)[0]['generated_text']
print("\n🗣️ AGI Voice Agent Reply:\n", response)
```
