```python
# AGS 014 ‚Äì Scientific Cause & Effect Reasoning

# üéØ Goal: Help AGS understand causal relationships in science (e.g., physics, biology, chemistry).

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
from wordcloud import WordCloud

# ‚úÖ Load dataset (upload to Kaggle as ags_014_scientific_causality.json)
with open('/kaggle/input/ags-014-scientific-causality/ags_014_scientific_causality.json', 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data)
df.head()




üß™ Count of Scientific Domains

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='domain', palette='Set2')
plt.title("Scientific Domains in Dataset")
plt.xlabel("Domain")
plt.ylabel("Count")
plt.tight_layout()
plt.show()




‚òÅÔ∏è Word Cloud of Causes

text = ' '.join(df['cause'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Common Cause Phrases")
plt.show()




üß† Sample Reasoning Example

sample = df.sample(1).iloc[0]

print("üåç Domain:", sample['domain'])
print("üîÑ Cause:", sample['cause'])
print("‚û°Ô∏è Effect:", sample['effect'])
print("üß† AGS Reasoning:", sample['ags_reasoning'])




‚úÖ Summary

- Teaches AGS how natural systems respond to changes
- Useful in simulation, robotics, tutoring systems




üìÅ Sample JSON Data

Save as: ags_014_scientific_causality.json

[
  {
    "id": "sci_001",
    "domain": "physics",
    "cause": "A ball is dropped from a height",
    "effect": "The ball accelerates downward due to gravity",
    "ags_reasoning": "Gravity causes the object to accelerate at 9.8 m/s¬≤ toward the Earth."
  },
  {
    "id": "sci_002",
    "domain": "biology",
    "cause": "A person is infected by a virus",
    "effect": "The person develops a fever and immune response",
    "ags_reasoning": "The immune system detects viral presence and raises body temperature to hinder viral replication."
  }
]
```
