```python
# AGS 004 ‚Äì Emotional & Social Reasoning Dataset

# üß† Purpose: Teach AGS agents to understand and reason about human emotional and social situations.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import json

# ‚úÖ Load dataset from URL or local
# If you're running this on Kaggle, upload `ags_004_emotion_social_reasoning.json`
with open('/kaggle/input/ags-004-emotion-social-reasoning/ags_004_emotion_social_reasoning.json', 'r') as f:
    data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(data)

# üîç Preview
df.head()




üìä Emotion Distribution

plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='emotion', order=df['emotion'].value_counts().index, palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Distribution of Emotions')
plt.tight_layout()
plt.show()




‚òÅÔ∏è Word Cloud of Situations

text = ' '.join(df['situation'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Situations')
plt.show()




üß† Reasoning Type Count

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='reasoning_type', palette='Set2')
plt.title('Reasoning Type Frequency')
plt.show()




ü§ñ Sample Prompt for AGS Agent

sample = df.sample(1).iloc[0]

print("üß© Situation:", sample['situation'])
print("‚ùì Question:", sample['question'])
print("‚úÖ Ideal Response:", sample['answer'])
print("üí° Emotion Involved:", sample['emotion'])
print("üîç Reasoning Type:", sample['reasoning_type'])




‚úÖ Summary

This notebook is part of the AGS (Artificial General System) learning modules.

- Teaches social and emotional reasoning
- Prepares AGS agents for human-aligned interaction
- Based on real-world like emotional scenarios

üîó Dataset Path: `datasets/ags_004_emotion_social_reasoning.json`




‚úÖ To Upload Dataset

Save a JSON file like this locally and upload to Kaggle:

[
  {
    "id": "er_001",
    "situation": "A child drops their ice cream cone.",
    "emotion": "Sadness",
    "question": "Why might the child feel sad?",
    "answer": "Because they were enjoying the ice cream and now it‚Äôs gone.",
    "reasoning_type": "emotional reasoning"
  },
  {
    "id": "er_002",
    "situation": "A friend ignores your text.",
    "emotion": "Confusion",
    "question": "Why might you feel confused?",
    "answer": "Because you expected a reply but didn't get one.",
    "reasoning_type": "social inference"
  }
]

Save as: ags_004_emotion_social_reasoning.json
```
