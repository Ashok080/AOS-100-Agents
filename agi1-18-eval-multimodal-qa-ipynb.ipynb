```python
# AGI1_18_eval_multimodal_qa.ipynb
# Multimodal question answering (simulated for Kaggle)

!pip install transformers torchaudio --quiet
from transformers import pipeline

# Load models
text_llm = pipeline("text-generation", model="distilgpt2")

# Simulate vision input (you'd use CLIP or BLIP for real)
image_caption = "A person holding an umbrella while it's raining."

# Simulate audio transcription (you'd use Whisper for real)
audio_transcript = "The speaker is laughing loudly."

# Multimodal prompt
multimodal_prompt = f"""
Image description: {image_caption}
Audio transcript: {audio_transcript}

Q: What is happening in this scene?
A:"""

# AGI reply
response = text_llm(multimodal_prompt, max_length=100, do_sample=True)[0]['generated_text']
print("🧠 AGI Multimodal Answer:\n")
print(response.strip())
```
