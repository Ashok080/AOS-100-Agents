```python
# AGI1_10_rag_memory_engine.ipynb
# Build a small RAG system using Hugging Face + in-memory FAISS

!pip install transformers faiss-cpu sentence-transformers --quiet

from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from transformers import pipeline

# ✅ 1. Load embedder + QA model
embedder = SentenceTransformer("all-MiniLM-L6-v2")
qa_model = pipeline("text-generation", model="distilgpt2")

# ✅ 2. Example knowledge base
documents = [
    "The brain has about 86 billion neurons.",
    "CLIP is a vision-language model developed by OpenAI.",
    "Whisper is used for speech-to-text conversion.",
    "RAG stands for Retrieval-Augmented Generation.",
    "AGI stands for Artificial General Intelligence."
]

# ✅ 3. Create document embeddings
doc_embeddings = embedder.encode(documents)
index = faiss.IndexFlatL2(doc_embeddings[0].shape[0])
index.add(np.array(doc_embeddings))

# ✅ 4. Ask a question
query = "What is Whisper used for?"
query_vector = embedder.encode([query])
D, I = index.search(np.array(query_vector), k=1)

# ✅ 5. Retrieve top document + generate answer
retrieved = documents[I[0][0]]
prompt = f"Q: {query}\nContext: {retrieved}\nA:"
answer = qa_model(prompt, max_length=50, do_sample=True)[0]['generated_text']

print("🔍 Top Match:", retrieved)
print("\n🧠 AGI Answer:\n", answer)
```
