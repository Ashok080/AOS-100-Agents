```python
# AGI1_08_audio_model_train.ipynb
# Load Whisper (tiny) for speech-to-text inference

!pip install transformers torchaudio --quiet

from transformers import WhisperProcessor, WhisperForConditionalGeneration
import torchaudio

# ‚úÖ Load Whisper tiny model
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-tiny.en")
processor = WhisperProcessor.from_pretrained("openai/whisper-tiny.en")

# ‚úÖ Load example audio (from torchaudio built-in dataset)
# This will download ~1 example if not cached
sample_audio, sample_rate = torchaudio.load("https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/sample1.flac")

# ‚úÖ Prepare input
input_features = processor(sample_audio[0], sampling_rate=sample_rate, return_tensors="pt").input_features

# ‚úÖ Generate transcription
predicted_ids = model.generate(input_features)
transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]

print("üîä Transcription:", transcription)
```
