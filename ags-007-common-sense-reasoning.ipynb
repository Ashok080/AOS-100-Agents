```python
# AGS 007 ‚Äì Common Sense Reasoning Dataset

# üß† Purpose: Train AGS agents to apply basic everyday knowledge and expectations.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import json

# ‚úÖ Load dataset from uploaded file (upload on Kaggle manually)
with open('/kaggle/input/ags-007-common-sense-reasoning/ags_007_common_sense_reasoning.json', 'r') as f:
    data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(data)

# üîç Preview
df.head()




üìä Reasoning Type Distribution

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='reasoning_type', palette='viridis')
plt.title('Distribution of Common Sense Reasoning Types')
plt.xlabel('Reasoning Type')
plt.ylabel('Count')
plt.show()




‚òÅÔ∏è Word Cloud of Situations

text = ' '.join(df['situation'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Everyday Situations')
plt.show()




üß† Sample Common Sense Reasoning Prompt

sample = df.sample(1).iloc[0]

print("üìå Situation:", sample['situation'])
print("‚ùì Question:", sample['question'])
print("‚úÖ Answer:", sample['answer'])
print("üß† Reasoning Type:", sample['reasoning_type'])




‚úÖ Summary

This notebook helps AGS agents build practical intelligence by applying obvious truths and lived experiences.

üìò Focus:
- Understand everyday logic
- Improve real-world AI usability
- Core for social and situational awareness

üîó Dataset Path: ags_007_common_sense_reasoning.json

üìÅ Dataset Upload Format Example

To use locally, save a JSON like this:

Filename: ags_007_common_sense_reasoning.json

[
  {
    "id": "cs_001",
    "situation": "It‚Äôs snowing heavily outside.",
    "question": "Should you wear sandals or boots?",
    "answer": "You should wear boots.",
    "reasoning_type": "common sense"
  },
  {
    "id": "cs_002",
    "situation": "A person is sleeping in a dark room.",
    "question": "What will happen if you turn on the light suddenly?",
    "answer": "They may wake up or get disturbed.",
    "reasoning_type": "common sense"
  }
]
```
