```python
# AGS 065 ‚Äì Image Captioning using BLIP

!pip install transformers torch torchvision timm pillow -q
from PIL import Image
import requests
from transformers import BlipProcessor, BlipForConditionalGeneration

# Load processor and model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Sample image
img_url = "https://huggingface.co/datasets/nateraw/image-captioning-demo/resolve/main/merlion.png"
image = Image.open(requests.get(img_url, stream=True).raw).convert('RGB')

# Preprocess
inputs = processor(images=image, return_tensors="pt")

# Generate caption
out = model.generate(**inputs)
caption = processor.decode(out[0], skip_special_tokens=True)

# Output
print("üñºÔ∏è Image Caption:")
print(caption)
image.show()
```
