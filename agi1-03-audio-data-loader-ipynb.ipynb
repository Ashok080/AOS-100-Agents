```python
# AGI1_03_audio_data_loader.ipynb
# Load AGI-scale audio datasets like LibriSpeech using Hugging Face

!pip install datasets torchaudio --quiet

from datasets import load_dataset
from itertools import islice
import torchaudio
import IPython.display as ipd

# ‚úÖ Load LibriSpeech audio dataset using streaming
dataset = load_dataset("librispeech_asr", "clean", split="train.100", streaming=True)

# ‚úÖ Sample 3 audio entries
sample = list(islice(dataset, 3))

# ‚úÖ Play and preview audio + text
for i, entry in enumerate(sample):
    print(f"üîä Sample {i+1}:")
    print("Transcript:", entry['text'])
    print("Speaker ID:", entry['speaker_id'])
    print("Audio Path:", entry['audio']['path'])
    print("="*40)
    
    # Stream and play audio
    audio = entry['audio']
    array, sr = torchaudio.load(audio['path'])
    ipd.display(ipd.Audio(array.numpy(), rate=sr))
```
