```python
# AGS 013 ‚Äì Visual Scene Understanding and Description

# üéØ Goal: Train AGS to describe complex scenes, detect multiple objects, and infer interactions between them.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
from wordcloud import WordCloud

# ‚úÖ Load dataset (upload to Kaggle as ags_013_visual_scenes.json)
with open('/kaggle/input/ags-013-visual-scenes/ags_013_visual_scenes.json', 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data)
df.head()




üìä Scene Types Distribution

plt.figure(figsize=(10,5))
sns.countplot(data=df, x='scene_type', palette='coolwarm')
plt.title('Types of Visual Scenes')
plt.xticks(rotation=45)
plt.xlabel("Scene Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()




‚òÅÔ∏è Word Cloud of Descriptions

text = ' '.join(df['description'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Scene Description Word Cloud')
plt.show()




üîç Random Scene Description

sample = df.sample(1).iloc[0]

print("üñºÔ∏è Scene Type:", sample['scene_type'])
print("üìù Description:", sample['description'])
print("üîç Key Objects:", ', '.join(sample['key_objects']))
print("ü§ñ AGS Inference:", sample['ags_inference'])




‚úÖ Summary

Use case:
- Vision + language alignment in AGI
- Describe images for accessibility
- Build spatial and situational awareness




üìÅ Sample JSON Data

Save as: ags_013_visual_scenes.json

[
  {
    "id": "vs_001",
    "scene_type": "park",
    "description": "Children are playing on a swing set while parents sit on nearby benches. A dog runs across the grass.",
    "key_objects": ["children", "swings", "parents", "dog", "grass"],
    "ags_inference": "It's a family-friendly area, possibly a weekend or holiday outing."
  },
  {
    "id": "vs_002",
    "scene_type": "kitchen",
    "description": "A man is chopping vegetables on a cutting board next to a stove. Steam rises from a pot.",
    "key_objects": ["man", "vegetables", "cutting board", "stove", "pot"],
    "ags_inference": "A meal is being prepared, possibly dinner in progress."
  }
]
```
