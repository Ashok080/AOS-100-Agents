```python
# AGI1_02_image_data_loader.ipynb
# Load AGI-scale image datasets like CIFAR or LAION using streaming

!pip install datasets torchvision --quiet

from datasets import load_dataset
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image

# ✅ Load CIFAR-10 from Hugging Face
dataset = load_dataset("cifar10", split="train")

# ✅ Preview first 5 images and labels
for i in range(5):
    img = Image.fromarray(dataset[i]['img'])
    label = dataset[i]['label']
    
    plt.imshow(img)
    plt.title(f"Label: {label}")
    plt.axis('off')
    plt.show()
```

```python
# Load a small LAION subset (if internet is fast enough)
laion = load_dataset("laion/laion400m", split="train", streaming=True)

# Stream sample
from itertools import islice
sample_laion = list(islice(laion, 5))

for item in sample_laion:
    print("Text:", item.get("TEXT", ""))
    print("URL:", item.get("URL", ""))
    print("="*50)
```
