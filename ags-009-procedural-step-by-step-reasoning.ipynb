```python
# AGS 009 ‚Äì Procedural & Step-by-Step Reasoning

# üõ†Ô∏è Goal: Teach AGS agents how to follow and reason through step-by-step instructions

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
from wordcloud import WordCloud

# ‚úÖ Load dataset (upload this file to Kaggle: ags_009_procedural_reasoning.json)
with open('/kaggle/input/ags-009-procedural-reasoning/ags_009_procedural_reasoning.json', 'r') as f:
    data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(data)

# üîç Preview
df.head()




üìä Distribution of Task Types

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='reasoning_type', palette='coolwarm')
plt.title('Step-by-Step Reasoning Types')
plt.xlabel('Reasoning Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()




‚òÅÔ∏è Word Cloud of Procedural Tasks

text = ' '.join(df['situation'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Procedural Situations')
plt.show()




üß™ Try a Sample Prompt

sample = df.sample(1).iloc[0]

print("üìù Situation:", sample['situation'])
print("‚ùì Question:", sample['question'])
print("‚úÖ Answer:", sample['answer'])
print("üìö Reasoning Type:", sample['reasoning_type'])




‚úÖ Summary

This dataset helps AGS agents practice procedural reasoning across real-world task sequences.

üß† Application Areas:
- Home assistant AI
- Cooking/assembly robots
- Helpdesk agents

üìÅ Dataset file: ags_009_procedural_reasoning.json




üìÅ Example Dataset Format

Save your JSON like this:

Filename: ags_009_procedural_reasoning.json

[
  {
    "id": "proc_001",
    "situation": "You want to make a cup of tea.",
    "question": "What steps do you follow?",
    "answer": "Boil water, place tea bag in cup, pour hot water, let it steep, then drink.",
    "reasoning_type": "procedural reasoning"
  },
  {
    "id": "proc_002",
    "situation": "You need to assemble a bookshelf.",
    "question": "What should you do first?",
    "answer": "Read the instruction manual.",
    "reasoning_type": "step initiation"
  }
]


```
