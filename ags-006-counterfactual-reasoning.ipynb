```python
# AGS 006 ‚Äì Counterfactual Reasoning Dataset

# üß† Purpose: Train AGS agents to understand "what if" reasoning and alternate outcomes.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import json

# ‚úÖ Load dataset from uploaded file (upload on Kaggle manually)
with open('/kaggle/input/ags-006-counterfactual-reasoning/ags_006_counterfactual_reasoning.json', 'r') as f:
    data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(data)

# üîç Preview
df.head()




üìä Reasoning Type Distribution

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='reasoning_type', palette='cool')
plt.title('Distribution of Counterfactual Reasoning Types')
plt.xlabel('Reasoning Type')
plt.ylabel('Count')
plt.show()




‚òÅÔ∏è Word Cloud of Scenarios

text = ' '.join(df['situation'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Counterfactual Situations')
plt.show()




üß† Sample Counterfactual Reasoning Prompt

sample = df.sample(1).iloc[0]

print("üß© Situation:", sample['situation'])
print("‚ùì Counterfactual Question:", sample['question'])
print("‚úÖ Reasoned Answer:", sample['answer'])
print("üåÄ Type:", sample['reasoning_type'])




‚úÖ Summary

This notebook trains AGS agents to explore hypothetical outcomes and "what if" scenarios.

üìò Focus:
- Learn from alternate pasts
- Critical for decision modeling
- Enhances foresight and imagination

üîó Dataset Path: ags_006_counterfactual_reasoning.json




üìÅ Dataset Upload Format Example

To use locally, save a JSON like this:

Filename: ags_006_counterfactual_reasoning.json

[
  {
    "id": "cf_001",
    "situation": "It rained yesterday and the picnic was cancelled.",
    "question": "What would have happened if it hadn't rained?",
    "answer": "The picnic would have taken place as planned.",
    "reasoning_type": "counterfactual"
  },
  {
    "id": "cf_002",
    "situation": "Lena forgot to study and failed the test.",
    "question": "What if Lena had studied properly?",
    "answer": "She likely would have passed the test.",
    "reasoning_type": "counterfactual"
  }
]
```
