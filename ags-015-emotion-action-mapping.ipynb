```python
# AGS 015 ‚Äì Emotion + Action Mapping

# üéØ Goal: Help AGS understand what actions people take based on emotions.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
from wordcloud import WordCloud

# ‚úÖ Load dataset (upload to Kaggle as ags_015_emotion_action.json)
with open('/kaggle/input/ags-015-emotion-action/ags_015_emotion_action.json', 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data)
df.head()




üìä Count of Emotions

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='emotion', palette='coolwarm')
plt.title("Distribution of Emotions")
plt.xlabel("Emotion")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()




‚òÅÔ∏è Word Cloud of Actions

text = ' '.join(df['action'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Common Actions Taken Based on Emotions")
plt.show()




üß† Example Mapping

sample = df.sample(1).iloc[0]

print("üí¨ Situation:", sample['situation'])
print("üíì Emotion:", sample['emotion'])
print("üîß Action Taken:", sample['action'])
print("üß† AGS Interpretation:", sample['ags_reasoning'])




‚úÖ Summary

- Links emotional context to likely behavioral response
- Enables AGS to act or suggest human-like decisions




üìÅ Sample JSON Data

Save as: ags_015_emotion_action.json

[
  {
    "id": "ea_001",
    "situation": "Someone shouts at a person in public.",
    "emotion": "Embarrassment",
    "action": "The person avoids eye contact and walks away quickly.",
    "ags_reasoning": "Embarrassment causes people to withdraw or avoid further attention."
  },
  {
    "id": "ea_002",
    "situation": "A child receives praise from a teacher.",
    "emotion": "Pride",
    "action": "The child smiles and tells their parents about the praise.",
    "ags_reasoning": "Pride leads to sharing accomplishments and feeling validated."
  }
]
```
