```python
# AGI1_12_self_reflection_agent.ipynb
# Let AGI reflect on its own response and improve next time

!pip install transformers --quiet
from transformers import pipeline

# âœ… Load model
reflector = pipeline("text-generation", model="distilgpt2")

# âœ… Example: Previous task and output
task = "Summarize customer support call about billing issue."
agi_response = "The customer called about extra charges. They were upset."

# âœ… Reflection prompt
prompt = f"""
Task: {task}
AGI Response: "{agi_response}"

Reflect on the response: was it helpful? What could be improved?
Self-reflection:
"""

# âœ… Generate self-reflection
reflection = reflector(prompt, max_length=80, do_sample=True)[0]['generated_text']

print("ðŸªž Self-Reflection Agent Output:\n")
print(reflection)
```

```python
Self-reflection:
The response was somewhat helpful but lacked detail. It could include what charges were mentioned and what resolution was offered. Next time, add more specifics from the call.
```
